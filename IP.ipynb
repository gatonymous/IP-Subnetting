### MASCARA

```python
def Mascara(byt,n_bits): # c=1 B=2 A=3
    bits_ = (32-byt*8)*'1'+n_bits*'1'+(byt*8-n_bits)*'0'

    Bits_ = []    
    # en decimal
    dec_ = []    
    k = 0
    for j in range(4):            
        Bits_.append(bits_[k:k+8])        
        # binario a decimal
        dec_.append(int(bits_[k:k+8], 2))       
        k += 8
    #print('MASCARA:',dec_)
    ms = '.'.join(str(x) for x in dec_) #array to string
    print('MASCARA:',colored(ms,'blue')) 
```

### SUB-REDES

```python
def SubRedes(ip, nroSR):
    n_bits = m.ceil(m.log2(nroSR)) # bits para sub-red

    # Identificar clase de red
    Ip = ip.split('.')
    if (int(Ip[0]) >= 0 and int(Ip[0]) <= 127):
        print('CLASE A')
        print('mascara:','255.0.0.0')
        byt = 3
        Mascara(byt,n_bits) # c=1 B=2 A=3
        print(n_bits,'bits','|','2^'+str(byt*8-n_bits)+'=',2**(byt*8-n_bits),'Host por red')
        print('')

    if (int(Ip[0]) >= 128 and int(Ip[0]) <= 191):
        print('CLASE B')
        print('mascara:','255.255.0.0')
        byt = 2
        Mascara(byt,n_bits) # c=1 B=2 A=3
        print(n_bits,'bits','|','2^'+str(byt*8-n_bits)+'=',2**(byt*8-n_bits),'Host por red')
        print('')

    if (int(Ip[0]) >= 192 and int(Ip[0]) <= 223):
        print('CLASE C')
        print('mascara:','255.255.255.0')
        byt = 1
        Mascara(byt,n_bits) # c=1 B=2 A=3
        print(n_bits,'bits','|','2^'+str(byt*8-n_bits)+'=',2**(byt*8-n_bits),'Host por red')
        print('')
        
    # generar bits
    Id = (byt*8 - n_bits)*'0'
    Brk = (byt*8 - n_bits)*'1'

    # generear binarios para sub-red
    for i in range(0,2**n_bits):
        temp = format(i, "b")
        #temp = "{0:b}".format(i)
        #print(bin(i))
        getbinary = lambda x, n: format(x, 'b').zfill(n) # retornar binario longitud n
        bits_0 = getbinary(i, n_bits) + Id
        bits_1 = getbinary(i, n_bits) + Brk

        # 1's y 0's
        bits_a = []
        bits_b = []
        # en decimal
        dec_0 = []
        dec_1 = []
        k = 0
        for j in range(byt):            
            bits_a.append(bits_0[k:k+8])
            bits_b.append(bits_1[k:k+8])
            # binario a decimal
            dec_0.append(int(bits_0[k:k+8], 2))
            dec_1.append(int(bits_1[k:k+8], 2))
            k += 8

        #print(bits_a,dec_0)
        #print(bits_b,dec_1)
        
        # Mostrar en colores 
        PrintBits(byt, n_bits, bits_a)

        r = '.'.join(str(x) for x in dec_0) #array to string
        print('>ID '+str(i+1),colored(r,'blue'))

        PrintBits(byt, n_bits, bits_b)

        s = '.'.join(str(x) for x in dec_1) #array to string
        print('>BRD'+str(i+1),colored(s,'blue'))    

        
        print('------------------------------------------')
        
 
```
